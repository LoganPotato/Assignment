```python
class Solution:
    def fizzBuzz(self, n):
    if n < 1 or n > 10000:
      return

    list = []
    for i in range(1, n):
        if i % 3 == 0 and i % == 0:
            list.append("FizzBuzz")
        elif i % 3 == 0:
            list.append("Fizz")
        elif i % 5 == 0:
            list.append("Buzz")
        else:
            list.append(i)
    return list
```

## 2. Peer Review & Draft Code

I looked at this LeetCode solution under the solution tab: https://leetcode.com/problems/reverse-integer/solutions/7261432/clean-simple-beats-94-easy-to-understand-tm4l

From this I noticed that they were using a very nice way to check if the number had any negatives but they were using slices to get the numbers. I took this idea and formed my own version using if statements and for loops and creating an empty string list to then put those numbers in there for the reverse order.

```python
class Solution:
    def reverse(self, x):

        number = str(x)
        string_list = ""
        position = 0
        if number[0] == "-":

          for i in range(len(number) - 1,position -1,-1):
            string_list + number[i]
        if position == 1:
            string_list = "-" + string_list
        reversal = int(string_list)
        if reversal < 2**31 - 1 or reversal > -2**31:
            return 0

        return reversal
```

#  **Problem 4 | Two Sum**




## 1. Understanding & Breakdown


This problem is asking for us to take an array and take any 2 numbers from the array that will add up to the target number. We cannot use the same number twice and there is only one answer for each given array. We will have to compare 2 numbers using loops to get the desired outcome.

Plan:

1. First we need to create multiple loops for it to go through the entire list of numbers, the array can be any length so we need to use the len function
2. One loop will start from position 0 the other will start from position 1
3.  The loops will compare 2 different numbers to see if they add to the target value
4. If the target value is reached the 2 numbers are returned, if not the positions of each number will keep going to the very end of the array

## 2. Peer Review & Draft Code

I reviewed this solution on LeetCode in the solutions tab:
https://leetcode.com/problems/two-sum/solutions/3619262/3-methods-c-java-python-beginner-friendl-x595

The "len" function here is very helpful aswell as you don't know how big the array is or the exact length value so this will help it for any size of array.


```python
class Solution(object):
    def twoSum(self, nums, target):
        i = 0
        j = 1
        while i < len(nums):
            while j < len(nums):
                if nums[i] + nums[j] == target:
                    return [i,j]
                j + 1
            i + 1
```

#  **Problem 5 | Valid Palindrome**




## 1. Understanding & Breakdown

This problem asks us to take all of the letters to lowercase, remove all non-alphanumerican values and it is a palindrome. We then need to reverse the given phrase/input and see if it is identical when read from left to right and right to left.

Plan:

1. First we need to make the phrase that is inputted all lowercase
2. Make an empty string list so that we can put the reversed phrase into it and compare at the end
3. We need to loop using the range and len of the list so that we can split it later
4. We then need to remove all non-alphanumeric numbers, if the value is fine, then it will be added to the string list
5. After all of the iterations we then need to compare both the original list and reversed list and see if it is a palindrome

## 2. Peer Review & Draft Code

I used this solution in LeetCode under the solution tab:
https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-mdpk

From viewing this solution I saw that they used a built in function called isalnum():. This function on it's own checks to see if it is a alphanumeric number so that this can easily help reverse the phrase and I was able to use this to my advantage to filter out characters I didn't need in the new string list. I also saw they used "[::-1]" I saw this manytimes in previous problems and decided to use this to easily reverse the entire order in one attempt rather than having loops like I did before to reverse it.

```python
class Solution:
    def isPalindrome(self, s):
        s = s.lower()
        string_list = ""
        temp = ""

        for char in s:
            for i in range(len(s)):
              if char.isalnum():
                  string_list += char
              else:
                  temp += char
                  string_list += temp


        return string_list  == string_list[::-1]
```

## 3. Final Correct Code




```python
class Solution:
    def isPalindrome(self, s):
        s = s.lower()
        string_list = ""

        for char in s:
            if char.isalnum():
                string_list += char

        return string_list  == string_list[::-1]
```
