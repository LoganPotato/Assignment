#  Problem 1 | Valid Parentheses (Stacks)





## 1. Understanding & Breakdown

The problems gives us a string that only contains a normal bracket, square bracket or curly bracket. We need to see if the string is valid based off 3 cases. If the begginning bracket closes with it's same bracket, open brackets need to be closed in the right order and the closed bracket matches with it's respective open bracket.

Plan:

1. We'll need to create an empty stack first and we can use this to keep track of the brackets used in the string
2. We then will have a loop go through the string
3. Everytime an open bracket shows up we push it into the stack
4. If we find a closed bracket we need to check the stack, if it's empty then it is false, if it's not empty we then need to pop the top of the stack
5. If the bracket we popped doesn't match the closed bracket then it's false
6. If it does match the loop will continue, at the end the stack should be empty which tells us all of the brackets had their match and we can return true


## 2. Peer Review & Draft Code

This is the solution I viewed in LeetCode: https://leetcode.com/problems/valid-parentheses/solutions/7024027/beginner-freindlyjavacpythonjs-by-ashokv-sge4

Also explains how that using the stack can be helpful for making sure that the cases are met, as for example if the program runs and there are still elements within the stack then that tells us that some of the brackets weren't closed and we'd have to return false.



```python
class Solution:
    def isValid(self, s):
        stack = []

        for i in s:
            if i == "(" or i == "[" or i == "{"
                stack.append("i")
            if stack == []:
                return False

            top = stack.pop()

            if top == "(" and i != ")" or top == "[" and i != "]" or top == "{" and i != "}":
                return False

            if len(stack) == 0:
                return True
            else:
                return False
```

#  Problem 2 | Number of Students Unable to Eat Lunch (Queues)




## 1. Understanding & Breakdown

The problem tells us that there are 2 types of sandwiches used to feed students. All of the students are placed in a queue and choose between the sandwiches. The number of sandwiches will be the same as the number of students present and the sandwiches are placed in a stack. If a student at the front of the queue likes the sandwhich option that is at the top of the stack then they take it, if not the student then goes to the last position of the queue. We will then have to find the number of students that decided to not eat due to their preference of the sandwhich

Plan:

1. We need to have the students put into a queue and sandwiches into one as well
2. We will use a counter that will check the number of students who do not eat in a row
3. A loop is needed to check the students/sanwhiches to see if the student likes the sandwhich at the top of the stack
4. If they do the sandwhich is taken, if not the student is placed at the end of the queue
5. After all of the iterations we check to see if the counter now equals the amount of students, if they equal each other then this tells us that no one wants to eat the sandwhich at the top of the stack


## 2. Peer Review & Draft Code

This is the solution I looked at on LeetCode: https://leetcode.com/problems/number-of-students-unable-to-eat-lunch/solutions/7348672/simple-python-queue-beats-100-by-milburn-wzu0

I learned that you can use deque and .popleft to easily remove the student from the front of the queue and then to add them back at the end with the commonly known append() function. The Intuition and Apprach helped me to further understand the problem and how my code should attack this problem. It explains a very good process of popping the front, pushing to the back and the continuosly repeating it for the students who dont want a sandwhich.

```python
class Solution(object):
    def countStudents(self, students, sandwiches):
        student_queue = deque(students)
        sandwich_stack = deque(sandwiches)
        rotations = 0

        while student_queue and sandwich_stack:
            current_student = student_queue[0]
            top_sandwich = sandwich_stack[0]

            if current_student == top_sandwich:
                if len(student_queue) > 1:
                    student_queue.popleft()
                rotations = 0
            else:
                if len(student_queue) % 2 == 0:
                    student_queue.append(student_queue.popleft())
                rotations += 1

            if rotations == len(student_queue):
                break

        if student_queue:
            return len(student_queue)
        else:
            return 0
```

```python
class Solution(object):
    def isPowerOfThree(self, n):

        if (n <= 0) or (n % 3 != 0):
            return False
        if n == 1:
            return True

        while n % 3 == 0 and n != 1:
            n //= 3

        if n > 1:
            return self.isPowerOfThree(n)
        else:
            return True
```

## 1. Understanding & Breakdown

A problem arises where we need to find the defective version which ultimately leads to all the versions that come after the wrong one, are also all defective. We need to find the first bad version so that we can then find all of the bad versions that come after.

Plan:

1. First we make 2 variables, one will be the first and the other will be the second version
2. We then use a binary search and have a variable set to mid which will show us the middle version
3. We check the version that is before mid and see if it is bad or not
4. If the version before is not bad and the mid value is bad then the mid will be the first bad version
5. If the previous version is bad then we search the left side of the bad version that we found, then we take another mid point on the left side and continue to find the first bad value if it follows the step 4.
6. Once all the iterations the variable we chose at the start to be the first version will have been updated to where the first intiial bad version is
